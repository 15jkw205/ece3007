```python
#                       LAB-3 --- PROBLEM 2
# Name: Jakob West
# Date: 10/07/2024
# Class/Section #: ECE-3007-001

# Problem Statement: Write a single Jupyter Notebook (JN), with a 
# markup window with Assignment information, Name, date, class, 
# assignment name, problem statement, algorithm, copy of program code.
# Then, write an executable Python program in another window, in same 
# JN, to generate a list of 50 cosine data points, for x[n] given 
# below and plot the graph. X[n] = 2.0 cos( pi*n / 20)

# Algorithm:
# Import the random module

# Generate a list of 15 random integers between 0 and 10 (inclusive)

# Create a duplicate list and sort it in ascending order

# Print the original list of random integers
# Print the sorted list in ascending order


import random

# Generate a list of 15 random integers between 0 and 10 (inclusive)
random_list = [random.randint(0, 10) for _ in range(15)]

# Create a duplicate list and sort it in ascending order
sorted_list = sorted(random_list)

# Print both lists
print("Original list:", random_list)
print("Sorted list (Low-to-High):", sorted_list)
```
